```python
import csv
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files


drive.mount('/content/gdrive')

df = pd.read_csv('gdrive/My Drive/2022.01. ML Learning - Onix Internship/dataset.txt', ' ')

#print(df.itterows)

x = []
y = []
for i, row in df.iterrows():
  x.append(row['X'])
  y.append(row['Y'])
  
```

```python
maxY = np.amax(y)
maxPoz = np.argmax(y)

plt.annotate('Max', xy=(maxPoz, maxY), xycoords='data', xytext=(70, 95), textcoords='data', arrowprops=dict(facecolor='m'))

plt.plot(x, y)
plt.grid(True)
plt.title('Liner Chart', fontsize=15)
plt.show()
```

```python
plt.step(x, y, 'g-o', where='mid')
plt.title('Step Chart', fontsize=16)
plt.show()
```

```python
plt.stem(x, y, use_line_collection=True, markerfmt='x')
plt.title('Stem Chart', fontsize=17)
plt.show()
```

```python
plt.bar(x, y, width=2)
plt.title('Bar Chart', fontsize=19)
plt.show()
```

```python
y10 = [i*2 for i in y]

plt.figure(figsize=(12, 20))
plt.subplot(3, 2, 1)
plt.plot(x, y, '-')
plt.title('First chart')

plt.subplot(2, 2, 2)
plt.plot(x, y10, 'o-')

plt.title('Second Chart with 2x diferencce', fontsize=18)

plt.show()
```
