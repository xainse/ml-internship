```python
#3. Знайти всі унікальні жанри і розбити по окремим дата фреймам. 
genres = df.iloc[:,4]
uniq_genres = genres.unique()

df_genres = []

for g in uniq_genres:
  df_genres.append(df.loc[df['Genre'].isin([g])])
```

```python
# for g in df_genres:
i = 0
filename = ''
drive.mount('/content/gdrive') 
for g in df_genres:
  df_genres[i] = g.sort_values('Year', axis=0, ascending=True)  # 4. Упорядкувати рядки відсотрувавши по Year
  df_genres[i] = df_genres[i].reset_index(drop=True) # 5. Зробити переіндексіцію
  # 6. Зберегти датафрейми пожанрово в файли в гугл драйв
  filename = df_genres[i]['Genre'][0] + '.csv'
  print(filename)
  df_genres[i].to_csv(dir_name+filename, mode='w+')
  
  i+=1
```
