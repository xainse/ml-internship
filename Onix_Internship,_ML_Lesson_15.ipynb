```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot
from google.colab import drive

drive.mount('/content/gdrive')

dir_name = '/content/gdrive/MyDrive/2022.01. ML Learning - Onix Internship (Serhii Kholin)/Lesson_015/'

# load the image

data1 = tf.io.read_file(dir_name + '01.jpg')
data2 = tf.io.read_file(dir_name + '02.jpg')

img_leaf = tf.image.decode_jpeg(data1)
img_flowers = tf.image.decode_jpeg(data2)

# variant 1
result = img_leaf + img_flowers
tf.keras.utils.save_img(dir_name + 'result.jpg', result)
pyplot.subplot(330 + 1)
pyplot.imshow(result)

# variant 2
result = img_leaf - img_flowers
tf.keras.utils.save_img(dir_name + 'result2.jpg', result)
pyplot.subplot(330 + 2)
pyplot.imshow(result)

# variant 3
result = img_leaf + 2*img_flowers
tf.keras.utils.save_img(dir_name + 'result3.jpg', result)
pyplot.subplot(330 + 3)
pyplot.imshow(result)
```
