Написать класс, выполняющий математическое описание криволинейных трапеций (информация о криволинейной трапеции должна включать левую и правую границы, а также уравнение ограничивающей кривой).

Предусмотреть метод, вычисляющий площадь криволинейной трапе-
ции, а также метод, выполняющий построение графика функции f(x) в за-
данном диапазоне x ε [a; b].

Проверить работоспособность кода на примере следующих данных:


```python
from scipy import interpolate
import scipy.integrate
import math
import numpy as np
import matplotlib.pyplot as plt


class LessonX():
  a: 0
  b: 0
  fx: 0

  res: 0

  def __init__(self, fx, limitA, limitB) :
    self.fx = fx
    self.a = limitA
    self.b = limitB

  def calcArea(self):
    result, err = scipy.integrate.quad(self.fx, self.a, self.b)
    return result

  def drawGraph(self):
    x = np.linspace(self.a, self.b, 120)
    y = self.fx(x)

    plt.plot(x, y)
    plt.show()
  


f  = lambda x:np.sin(x)+np.cos(2*x)
f2 = lambda x:x*np.tan(x)+2
f3 = lambda x:x**3-10*x**2 + 170


fun1 = LessonX(f, 0, 1.2)
area = fun1.calcArea()
print('Area for fun1 is: ', area)
fun1.drawGraph()



fun2 = LessonX(f2, 0.2, 0.8)
area = fun2.calcArea()
print('Area for fun2 is: ', area)

fun2.drawGraph()


fun3 = LessonX(f3, 0, 10)
area = fun3.calcArea()
print('Area for fun3 is: ', area)

fun3.drawGraph()
```
